```python
#Funciones del git hub
"""
Descripción del código: Función para calcular el factorial de un número.
Utiliza un ciclo for para multiplicar sucesivamente los números del 1 al n.
Desarrollador: Fernando Villegas
Fecha: 2026-04-2
"""

def factorial(n):
    """
    Calcula el producto de todos los enteros desde 1 hasta n.
    n! = 1 * 2 * 3 * ... * n
    """
    # Verificamos si el número es negativo
    if n < 0:
        return "Error: El factorial no está definido para números negativos."

    # El factorial de 0 y 1 es siempre 1
    if n == 0 or n == 1:
        return 1

    # Inicializamos el acumulador en 1 (porque es multiplicación)
    resultado = 1

    # El rango debe ir hasta n + 1 para incluir el número n
    for i in range(1, n + 1):
        resultado = resultado * i

    return resultado

# --- Pruebas del Programa ---
numero = 5
print(f"El factorial de {numero} es: {factorial(numero)}")

# Ejemplo rápido: 5! = 1 * 2 * 3 * 4 * 5 = 120
```

```python
"""
Descripción del código: Función recursiva que filtra números pares y los eleva al cuadrado.
Sustituye la lógica de 'map' y 'filter' mediante recursión pura sobre listas.
Desarrollador: Fernando Villegas
Fecha: 2026-04-2
"""

def procesar_recursivo(lista):
    """
    Recorre una lista de forma recursiva.
    Si el número es par, lo eleva al cuadrado y lo conserva.
    Si es impar, lo descarta.
    """
    # 1. Caso base: Si la lista está vacía, retornamos una lista vacía
    if not lista:
        return []

    # 2. Separamos el primer elemento del resto
    primero = lista[0]
    resto = lista[1:]

    # 3. Lógica de decisión (Paso recursivo)
    if primero % 2 == 0:
        # Si es par: cuadrado + resultado de procesar el resto
        return [primero ** 2] + procesar_recursivo(resto)
    else:
        # Si es impar: ignoramos el primero y procesamos solo el resto
        return procesar_recursivo(resto)

# --- Pruebas del Programa ---
numeros = [1, 2, 3, 4, 5, 6]
resultado = procesar_recursivo(numeros)

print(f"Lista original: {numeros}")
print(f"Pares al cuadrado: {resultado}")
```

```python
"""
Descripción del código: Función recursiva para sumar elementos de una lista.
Aplica el concepto de división del problema (cabeza y cola) para acumular el total.
Desarrollador: Fernando Villegas
Fecha: 2026-04-2
"""

def suma_recursiva(lista):
    """
    Suma todos los números de una lista sin usar ciclos (for/while).
    Caso base: Si la lista está vacía, la suma es 0.
    Paso recursivo: Primer elemento + suma_recursiva(resto de la lista).
    """
    # 1. Caso base (Indispensable para que no sea un ciclo infinito)
    if not lista:
        return 0

    # 2. Paso recursivo
    # lista[0] es el primer elemento
    # lista[1:] es la lista sin el primer elemento
    return lista[0] + suma_recursiva(lista[1:])

# --- Pruebas del Programa ---
numeros_examen = [5, 10, 15, 20]
resultado = suma_recursiva(numeros_examen)

print(f"La lista es: {numeros_examen}")
print(f"La suma total (recursiva) es: {resultado}")
```

```python
"""
Descripción del código: Función recursiva para invertir una cadena de texto.
Utiliza la técnica de segmentación de strings para reordenar los caracteres.
Desarrollador: Fernando Villegas
Fecha: 2026-04-2
"""

def invertir_cadena(texto):
    """
    Invierte un string de manera recursiva.
    Caso base: Si el texto está vacío o tiene un solo carácter, se retorna igual.
    Paso recursivo: Se toma el último carácter y se le suma la inversión del resto.
    """
    # 1. Caso base
    if len(texto) <= 1:
        return texto

    # 2. Paso recursivo
    # Retornamos el último carácter + la función llamada con el resto del texto
    # texto[-1] es el último carácter
    # texto[:-1] es todo el texto excepto el último carácter
    return texto[-1] + invertir_cadena(texto[:-1])

# --- Pruebas del Programa ---
palabra = "Algoritmia"
resultado = invertir_cadena(palabra)

print(f"Texto original: {palabra}")
print(f"Texto invertido: {resultado}")
```

```python
"""
Descripción del código: Función recursiva para encontrar el valor máximo.
Sustituye la función integrada max() comparando elementos de dos en dos
mediante llamadas recursivas.
Desarrollador: Fernando Villegas
Fecha: 2026-04-2
"""

def encontrar_maximo_recursivo(lista):
    """
    Encuentra el número mayor en una lista sin usar ciclos ni max().
    Caso base: Si la lista tiene un solo elemento, ese es el máximo.
    Paso recursivo: Compara el primer elemento con el máximo del resto.
    """
    # 1. Caso base: Una lista de un solo elemento
    if len(lista) == 1:
        return lista[0]

    # 2. Paso recursivo: Obtener el máximo de la "cola" (el resto de la lista)
    max_del_resto = encontrar_maximo_recursivo(lista[1:])

    # 3. Comparación (La lógica del acumulador/reduce)
    if lista[0] > max_del_resto:
        return lista[0]
    else:
        return max_del_resto

# --- Pruebas del Programa ---
mis_notas = [3.5, 4.8, 2.9, 5.0, 4.2]
mayor = encontrar_maximo_recursivo(mis_notas)

print(f"Lista de valores: {mis_notas}")
print(f"El valor más grande encontrado es: {mayor}")
```

```python
"""
Habilidades: funciones, ciclos, condiciones, listas
1. Ejercicio 1 – Análisis de una lista de temperaturas
Luis Fernando Villegas Ramirez
Descripción del código: Sistema de análisis de temperaturas diarias (24 horas).
Calcula promedio, valores extremos y frecuencia sobre la media.
Desarrollador: Fernando Villegas
Fecha: 2026-04-18
CC: 1117807385
3134771451
"""
# 1. Función para calcular el promedio
def promedio(lista):
    suma = sum(lista)
    return suma / len(lista)

# 2. Función para encontrar el máximo y mínimo
def extremos(lista):
    mayor = lista[0]
    menor = lista[0]

    for temp in lista:
        if temp > mayor:
            mayor = temp
        if temp < menor:
            menor = temp

    return mayor, menor

# 3. Función para contar cuántas veces superó el promedio
def dias_sobre_promedio(lista):
    valor_promedio = promedio(lista)
    contador = 0

    for temp in lista:
        if temp > valor_promedio:
            contador += 1

    return contador

# --- Implementación ---
# Creamos una lista manual de 24 horas (puedes cambiar estos valores)
temperaturas_hoy = [15, 14, 14, 13, 12, 12, 15, 18, 20, 22, 24, 25,
                     26, 26, 25, 24, 22, 20, 18, 17, 16, 16, 15, 14]

# Ejecutamos las funciones
p = promedio(temperaturas_hoy)
maximo, minimo = extremos(temperaturas_hoy)
sobre_p = dias_sobre_promedio(temperaturas_hoy)

# Resultados
print(f"El promedio del día fue: {p:.2f}°C")
print(f"La temperatura más alta fue {maximo}°C y la más baja {minimo}°C")
print(f"Horas que estuvieron por encima del promedio: {sobre_p}")
```

```python
"""
Descripción del código: Calculadora Nivel 2.
Implementa operaciones básicas y avanzadas (potencia, raíz, factorial e inversa).
La potencia y la raíz se basan en la lógica de multiplicación.
Desarrollador: Fernando Villegas
Fecha: 2026-04-18
"""

# --- 1. Funciones Básicas ---

def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    if b == 0:
        return "Error: División por cero"
    return a / b

# --- 2. Funciones que dependen de Multiplicación ---

def exponente(base, potencia):
    """Calcula la potencia usando la función multiplicar en un ciclo."""
    resultado = 1
    for _ in range(int(potencia)):
        resultado = multiplicar(resultado, base)
    return resultado

def raiz_cuadrada(n):
    """Calcula la raíz usando el operador de potencia (n ** 0.5)."""
    if n < 0:
        return "Error: Raíz negativa"
    return n ** 0.5

# --- 3. Funciones Adicionales ---

def factorial(n):
    """Calcula el factorial de un número usando un ciclo for."""
    if n < 0:
        return "Error: No existe factorial de negativos"
    resultado = 1
    for i in range(1, int(n) + 1):
        resultado = multiplicar(resultado, i) # Reutilizamos multiplicar
    return resultado

def inversa(n):
    """Retorna la inversa de un número (1/n)."""
    if n == 0:
        return "Error: El cero no tiene inversa"
    return dividir(1, n) # Reutilizamos dividir

# --- 4. Función de Operación (Orquestador) ---

def operacion(tipo, val1, val2=None):
    """
    Recibe el tipo de operación y los valores necesarios.
    Usa condiciones para decidir qué función llamar.
    """
    if tipo == "suma":
        return sumar(val1, val2)
    elif tipo == "resta":
        return restar(val1, val2)
    elif tipo == "multi":
        return multiplicar(val1, val2)
    elif tipo == "div":
        return dividir(val1, val2)
    elif tipo == "potencia":
        return exponente(val1, val2)
    elif tipo == "raiz":
        return raiz_cuadrada(val1)
    elif tipo == "factorial":
        return factorial(val1)
    elif tipo == "inversa":
        return inversa(val1)
    else:
        return "Operación no válida"

# --- Pruebas del Programa ---
print(f"Suma (5+3): {operacion('suma', 5, 3)}")
print(f"Potencia (2^3): {operacion('potencia', 2, 3)}")
print(f"Factorial de 5: {operacion('factorial', 5)}")
print(f"Inversa de 2: {operacion('inversa', 2)}")
print(f"Raíz de 16: {operacion('raiz', 16)}")
```

```python
"""
Descripción del código: Sistema de gestión de inventario.
Permite administrar productos (nombre, cantidad, precio) usando diccionarios.
Incluye funciones de agregar, eliminar, calcular totales y visualizar.
Desarrollador: Fernando Villegas
Fecha: 2026-04-18
"""

# --- 1. Función para agregar o actualizar productos ---
def agregar_producto(diccionario, nombre, cantidad, precio):
    """Añade un nuevo producto o actualiza uno existente."""
    diccionario[nombre] = {
        "cantidad": cantidad,
        "precio": precio
    }
    print(f"Producto '{nombre}' gestionado correctamente.")

# --- 2. Función para eliminar productos ---
def eliminar_producto(diccionario, nombre):
    """Elimina un producto del diccionario si existe."""
    if nombre in diccionario:
        del diccionario[nombre]
        print(f"Producto '{nombre}' eliminado.")
    else:
        print(f"Error: El producto '{nombre}' no existe en el inventario.")

# --- 3. Función para calcular el valor total del inventario ---
def calcular_valor_total(diccionario):
    """Recorre el diccionario para sumar el (precio * cantidad) de cada item."""
    total = 0
    for nombre in diccionario:
        datos = diccionario[nombre]
        subtotal = datos["cantidad"] * datos["precio"]
        total += subtotal
    return total

# --- 4. Función para mostrar el inventario de forma estética ---
def mostrar_inventario(diccionario):
    """Imprime todos los productos registrados."""
    print("\n--- ESTADO ACTUAL DEL INVENTARIO ---")
    if not diccionario:
        print("El inventario está vacío.")
    else:
        for nombre, datos in diccionario.items():
            print(f"Producto: {nombre} | Cantidad: {datos['cantidad']} | Precio: ${datos['precio']}")
    print("------------------------------------\n")

# --- Programa Principal ---

# Creacion del diccionario vacio
mi_inventario = {}

# Agregamos algunos productos
agregar_producto(mi_inventario, "Manzanas", 50, 1.5)
agregar_producto(mi_inventario, "Peras", 30, 2.0)
agregar_producto(mi_inventario, "Cereal", 10, 5.5)

# Mostramos cómo va el inventario
mostrar_inventario(mi_inventario)

# Calculamos el valor total
total_dinero = calcular_valor_total(mi_inventario)
print(f"Valor total del inventario: ${total_dinero}")

# Eliminamos un producto y volvemos a mostrar
eliminar_producto(mi_inventario, "Peras")
mostrar_inventario(mi_inventario)
```

```python
"""
Descripción del código: Sistema de gestión de notas escolares.
Calcula promedios simples, ponderados y determina el estado de aprobación.
Desarrollador: Fernando Villegas
Fecha: 2026-04-20
"""

# --- 1. Función Promedio Simple ---
def promedio_simple(lista_notas):
    """Suma las notas y divide por la cantidad de elementos."""
    return sum(lista_notas) / len(lista_notas)

# --- 2. Función Promedio Ponderado ---
def promedio_ponderado(lista_notas):
    """Calcula la nota final según pesos: 30%, 30% y 40%."""
    # Accedemos a cada nota por su índice en la lista
    n1 = lista_notas[0] * 0.30
    n2 = lista_notas[1] * 0.30
    n3 = lista_notas[2] * 0.40
    return n1 + n2 + n3

# --- 3. Función Mejor Estudiante ---
def determinar_mejor_estudiante(diccionario_notas):
    """Compara los promedios ponderados y retorna el nombre del mayor."""
    mejor_nota = -1
    mejor_estudiante = ""

    for nombre, notas in diccionario_notas.items():
        promedio_actual = promedio_ponderado(notas)
        if promedio_actual > mejor_nota:
            mejor_nota = promedio_actual
            mejor_estudiante = nombre

    return mejor_estudiante, mejor_nota

# --- 4. Función de Aprobación y Bonus ---
def mostrar_estado_aprobacion(diccionario_notas):
    """Muestra quién aprobó y envía mensaje de la profesora McGonagall."""
    print("\n--- ACTA DE NOTAS: CLASE DE TRANSFORMACIONES ---")
    for nombre, notas in diccionario_notas.items():
        final = promedio_ponderado(notas)

        # Clasificación
        if final >= 3.0:
            estado = "APROBADO"
            mensaje = "¡Felicidades! Ha pasado la clase de la Prof. McGonagall."
        else:
            estado = "REPROBADO"
            mensaje = "Lamentable. Debe esforzarse más en el próximo ciclo."

        print(f"Estudiante: {nombre} | Final: {final:.2f} | Estado: {estado}")
        print(f"Nota de McGonagall: {mensaje}\n")

# --- Programa Principal ---

# Datos iniciales (Corregí un par de typos en los nombres del ejercicio)
registro_notas = {
    "Harry": [3.8, 4.0, 4.2],
    "Ron": [3.2, 3.8, 2.8],
    "Hermione": [5.0, 5.0, 5.0],
    "Draco": [4.5, 4.2, 5.0],
    "Neville": [2.5, 3.0, 3.2]
}

# Ejecución de tareas
mejor_alumno, su_nota = determinar_mejor_estudiante(registro_notas)
print(f"🏆 El mejor promedio fue de {mejor_alumno} con {su_nota:.2f}")

mostrar_estado_aprobacion(registro_notas)
```

```python
"""
Descripción del código: Juego "Adivina el número" con niveles de dificultad.
Implementa lógica de intentos limitados, generación aleatoria y registro de historial.
Desarrollador: Fernando Villegas
Fecha: 2026-04-20
"""

import random

# --- 1. Función para configurar la dificultad ---
def elegir_dificultad():
    """Retorna la cantidad de intentos según el nivel elegido."""
    print("\n--- NIVELES DE DIFICULTAD ---")
    print("1. Fácil (15 intentos)")
    print("2. Medio (10 intentos)")
    print("3. Difícil (5 intentos)")

    opcion = input("Elige un nivel (1, 2 o 3): ")

    if opcion == "1":
        return 15
    elif opcion == "2":
        return 10
    else:
        return 5

# --- 2. Función para generar el número secreto ---
def generar_numero():
    """Genera un número entero aleatorio entre 1 y 100."""
    return random.randint(1, 100)

# --- 3. Función Principal del Juego ---
def jugar(intentos_totales):
    """Ejecuta la lógica del juego y retorna el resultado para el historial."""
    numero_secreto = generar_numero()
    intentos_restantes = intentos_totales
    ganó = False

    print(f"\n¡He pensado un número entre 1 y 100! Tienes {intentos_totales} intentos.")

    while intentos_restantes > 0:
        print(f"\nIntentos restantes: {intentos_restantes}")
        intento_usuario = int(input("¿Cuál crees que es el número?: "))

        if intento_usuario == numero_secreto:
            print(f"¡FELICIDADES! Adivinaste el número {numero_secreto}.")
            ganó = True
            break
        elif intento_usuario < numero_secreto:
            print("Pista: El número secreto es MAYOR.")
        else:
            print("Pista: El número secreto es MENOR.")

        intentos_restantes -= 1

    if not ganó:
        print(f"\nSe agotaron los intentos. El número era {numero_secreto}.")

    return f"Resultado: {'Ganó' if ganó else 'Perdió'} | Intentos usados: {intentos_totales - intentos_restantes}"

# --- Programa Principal ---

historial = []
continuar = "s"

while continuar.lower() == "s":
    intentos = elegir_dificultad()
    resultado_partida = jugar(intentos)

    # Guardar en el historial (lista)
    historial.append(resultado_partida)

    continuar = input("\n¿Quieres jugar otra vez? (s/n): ")

# Mostrar el historial al finalizar
print("\n--- HISTORIAL DE PARTIDAS ---")
for partida in historial:
    print(partida)
print("¡Gracias por jugar!")
```

```python
"""
Descripción del código: Simulación de una máquina dispensadora.
Gestiona un catálogo de 5 productos, procesa pagos, valida montos mínimos
y calcula el cambio (devuelta) para el usuario.
Desarrollador: Fernando Villegas
Fecha: 2026-04-20
"""

# --- 1. Definición del Inventario (Diccionario) ---
productos = {
    "Papas": 2500,
    "Gaseosa": 3000,
    "Galletas": 1500,
    "Chocolatina": 2000,
    "Agua": 1800
}

# --- 2. Función de Validación (Bonus) ---
def validar_pago(precio_producto, dinero_ingresado):
    """
    Compara el dinero ingresado con el precio.
    Retorna True si alcanza, False si falta dinero.
    """
    if dinero_ingresado < precio_producto:
        falta = precio_producto - dinero_ingresado
        print(f"Dinero insuficiente. Te faltan ${falta}.")
        return False
    return True

# --- 3. Función para Procesar la Compra ---
def procesar_compra(nombre_producto, dinero_ingresado):
    """
    Calcula la devuelta y muestra el mensaje final del producto.
    """
    precio = productos[nombre_producto]

    # Usamos la función del bonus para validar
    if validar_pago(precio, dinero_ingresado):
        cambio = dinero_ingresado - precio

        print("\n--- DESPACHANDO ---")
        print(f"Producto entregado: {nombre_producto}")

        if cambio > 0:
            print(f"Tu devuelta es: ${cambio}")
        else:
            print("Pago exacto. No hay devuelta.")
        print("¡Gracias por tu compra!")

# --- 4. Programa Principal ---

def maquina_dispensadora():
    print("--- BIENVENIDO A LA DISPENSADORA ---")
    # Mostramos los productos disponibles
    for nombre, precio in productos.items():
        print(f"- {nombre}: ${precio}")

    print("------------------------------------")
    seleccion = input("Escribe el nombre del producto que deseas: ").capitalize()

    # Verificar si el producto existe
    if seleccion in productos:
        try:
            pago = int(input(f"El precio es ${productos[seleccion]}. Ingresa el dinero: "))
            procesar_compra(seleccion, pago)
        except ValueError:
            print("Error: Por favor ingresa un valor numérico válido.")
    else:
        print("Error: Ese producto no está en la máquina.")

# Ejecutar la máquina
maquina_dispensadora()
```
