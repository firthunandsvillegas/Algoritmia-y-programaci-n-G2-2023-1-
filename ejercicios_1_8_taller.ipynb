```python
#1.2 Todos los números positivos que son divisibles por 10 y menores que n.
n=int(input('ingrese un numero mayor multiplo de 10: '))
numero = 0
while numero < n:
    if numero % 10 == 0 and numero >0:
        print(numero)
    numero += 1
```

```python
#2.1 La suma de todos los números pares entre 2 y 100 (ambos inclusive).

suma_pares = 0
numero = 2
while numero <= 100:
    suma_pares += numero
    numero += 2

print(f'La suma de todos los números pares entre 2 y 100 es: {suma_pares}')
```

```python
#2.2 La suma de todos los cuadrados entre 1 y 100 (ambos inclusive).
suma_cuadrados = 0
numero = 1
while numero <= 100:
    suma_cuadrados += numero * numero
    numero+=1
print(f'La suma de todos los cuadrados entre 1 y 100 es: {suma_cuadrados}')
```

```python
##2.3 La suma de todos los números impares entre a y b (ambos inclusive).
suma_impares = 0
z = 1

while z <= 9:
    if z % 2 != 0:
        suma_impares += z
    z += 1

print(f'la suma de los impares entre 1 y 9 es: {suma_impares}')
```

```python
#2.4 La suma de todos los dígitos impares de n.
n = "32255"
suma_impares = 0

for digito in n:
    d = int(digito)
    if d % 2 != 0:
        suma_impares += d

print(f'la suma de los impares de {n} es : {suma_impares}')
```

```python
#3.3 La suma de todos los dígitos impares de una entrada.
n = int(input("Ingrese un número: "))
suma_impares = 0
while n > 0:
    digito = n % 10
    if digito % 2 != 0:
        suma_impares += digito
    n //= 10

print(f'La suma de los dígitos impares es: {suma_impares}')
```

```python
#4. Si en un reloj digital registra las 6 AM, hacer sonar una alarma por 1 minuto a
# menos que se accione el botón de apagado.

while True:
    h = int(input('Ingrese la hora actual: '))

    if h == 6:
        print('Sonando alarma')

        b = int(input('Presione 1 para apagar: '))

        if b == 1:
            print('Alarma detenida')
        else:
            print('La alarma seguirá sonando por 60 segundos')

    else:
        print('No es hora de sonar la alarma')
```
